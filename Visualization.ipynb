```python
import json
```

```python
path = './result.json'
```

```python
with open(path, "r", encoding="utf-8") as f:
    result_dict = json.load(f)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
emcdr_loss = result_dict['EMCDR_Model_S2T']['loss']
epoch = range(len(emcdr_loss))
plt.plot(epoch, emcdr_loss)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('EMCDR Loss (S -> T)')
plt.show()
```

```python
def plot_helper(data_dict, model_name, keys, x_label, y_label, title):
    plt.figure(figsize=(10, 5))
    data = []
    plt.ylim([0, 1.])
    for key in keys:
        data.append((key, data_dict[model_name][key])) 
    epoch = range(1, len(data[0][1])+1)
    for datum in data:
        plt.plot(epoch, datum[1], label=datum[0])
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
model_name = 'EMCDR_Model_S2T'
keys = ['Hits@1', 'Hits@5', 'Hits@10', 'Hits@15']
x_label = 'Epoch'
y_label = 'Hits@K'
title = 'Hits@K Performance of EMCDR Model (S -> T)'
plot_helper(result_dict, model_name, keys, x_label, y_label, title)
```

```python
model_name = 'EMCDR_Model_S2T'
keys = ['NDCG@1', 'NDCG@5', 'NDCG@10', 'NDCG@15']
x_label = 'Epoch'
y_label = 'NDCG@K'
title = 'NDCG@K Performance of EMCDR Model (S -> T)'
plot_helper(result_dict, model_name, keys, x_label, y_label, title)
```
